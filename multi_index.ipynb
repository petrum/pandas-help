```python
import pandas as pd

def getData():
    df = pd.DataFrame(columns=['lastname', 'firstname', 'address', 'male'])
    df.set_index(['lastname', 'firstname'], inplace=True)
    #df.sort_index(inplace=True)
    df.sortlevel(inplace=True)
    df.ix[('Marginean', 'Maria'), :] = ['16 RCE', False]
    df.ix[('Marginean', 'Bogdan'), :] = ['16 RCE', True]    
    df.ix[('Mirabela', 'Maria'), :] = ['16 RCE', False]
    return df

df = getData()

df.loc['Marginean', 'Maria'].address = 'NYC1'
print(df)
df.loc['Marginean'].address = 'NYC2'
print(df)
df.loc[('Marginean', 'Maria'), 'address'] = 'NYC3'
print(df)
df.loc[('Marginean', 'Maria')].address = 'NYC4'
print(df)
df.ix[('Marginean', 'Maria')].address = 'NYC5'
print(df)
df.ix[['Marginean', 'Maria']].address = 'NYC6' # attention!!! it does nothing...
print(df)
df.ix['Marginean', 'Maria'].address = 'NYC7'
print(df)
df.ix['Marginean'].address = 'NYC8'
print(df)
df.ix[('Marginean')].address = 'NYC9'
# http://pandas.pydata.org/pandas-docs/stable/indexing.html#indexing-view-versus-copy
print(df)
df.loc[(slice(None), 'Maria'), 'address'] = 'NY10'
print(df)
idx = pd.IndexSlice
df.loc[idx[:, 'Maria'], 'address'] = 'NY11'
print(df)
idx = pd.IndexSlice
df.loc[idx[:, 'Maria'], idx['address']] = 'NY12'
print(df)
if ('Marginean', 'Maria') in df.index:
    #df = df.loc[df.index.drop(('Marginean', 'Maria'))]
    df.drop(('Marginean', 'Maria'), inplace=True)
df = df.loc[df.index.drop(('Marginean', 'Bogdan'))]
print(df)
```
